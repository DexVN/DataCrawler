# Reads data by book number.

```python
import json
import os
import re

# Load the JSON data from the file
with open('data/filtered_birth_data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Initialize a list to store filtered items
filtered_items = []

year_book = "2018"
# Book no
book_no = f"01/{year_book}"

print(book_no)
# Loop through each entry in the data
for entry in data:
    filtered_entry_items = [item for item in entry['items'] if item.get("bookNo") == book_no]
    if filtered_entry_items:
        filtered_items = filtered_entry_items

print(filtered_items)
```

# Read file name

```python
def read_file_names(folder_path):
    try:
        # Get the list of files in the directory
        files = os.listdir(folder_path)
        return files
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Define the folder path
folder_path = rf"C:\Users\duykhanh\Documents\Khai sinh\08_update\khai_sinh_{year_book}_bo_sung"

# Read the file names
file_names = read_file_names(folder_path)

# Print the file names
for file_name in file_names:
    print(file_name)
```

# Rename file

```python
def rename_files(filenames):
    for filename in filenames:
        # Match the pattern KS.YYYY.MM.YYYY-MM-DD.HH.pdf
        match = re.match(r'KS\.\d{4}\.\d{2}\.(\d{4})-(\d{2})-(\d{2})\.(\d+)\.jpeg', filename)
        if match:
            year = match.group(1)
            month = match.group(2)
            day = match.group(3)
            sequence = match.group(4)
            
            # Check sequence with a filter item data
            for item in filtered_items:
                if "numberNo" in item:
                    numberNo = item["numberNo"]
                    numberNo = numberNo.split("/")[0]

                    # Check match sequence and number no
                    if int(sequence) == int(numberNo):

                        # Get register date
                        register_data = item["registrationDateStr"]

                        # break date string to year, month, day
                        register_data = register_data.split("/")
                        register_year = register_data[2]
                        register_month = register_data[1]
                        register_day = register_data[0]
                        
                        # Rename file
                        new_filename = f"KS.{year}.01.{register_year}-{register_month}-{register_day}.{sequence}.jpeg"
                        os.rename(os.path.join(folder_path, filename), os.path.join(folder_path, new_filename))
                        print(f"Renaming {filename} to {new_filename}")
                        break  
                else:
                    print("Key 'numberNo' not found in item")

rename_files(file_names)
```

```python
# Delete files in folder if sequence not in filtered_items
def delete_files(filenames):
    for filename in filenames:
        # Match the pattern KS.YYYY.MM.YYYY-MM-DD.HH.pdf
        match = re.match(r'KS\.\d{4}\.\d{2}\.(\d{4})-(\d{2})-(\d{2})\.(\d+)\.jpeg', filename)
        if match:
            sequence = match.group(4)
            is_exist = False
            for item in filtered_items:
                if "numberNo" in item:
                    numberNo = item["numberNo"]
                    numberNo = numberNo.split("/")[0]

                    # Check match sequence and number no
                    if int(sequence) == int(numberNo):
                        is_exist = True
                        break
            if not is_exist:
                os.remove(os.path.join(folder_path, filename))
                print(f"Deleted {filename}")

delete_files(file_names)
```
