```python
import requests
import json
import time
# Replace with your API endpoint
url = 

# Define the payload
payload = {
    "birthRegistrationForm.status": -1,
    "birthRegistrationForm.inputDateFrom": "",
    "birthRegistrationForm.inputDateTo": "",
    "birthRegistrationForm.numberNo": "",
    "birthRegistrationForm.bookNo": "01/2002",
    "birthRegistrationForm.pageNo": "",
    "birthRegistrationForm.rpFullName": "",
    "birthRegistrationForm.rpBirthDate": "",
    "birthRegistrationForm.rpGender": -2,
    "birthRegistrationForm.registrationDateFrom": "",
    "birthRegistrationForm.registrationDateTo": "",
    "birthRegistrationForm.registrationPlaceStr": -1,
    "q": "*",
    "start": 0,
    "count": 1000,
    "startval": 0
}

# Define the cookies
cookies = {
    "JSESSIONID": "",
    "SRVNAME": ";"
}


all_data = []

# Loop through each year from 1999 to 2019
for year in range(1990, 2020):
    # Define the payload for each year
    payload = {
        "birthRegistrationForm.status": -1,
        "birthRegistrationForm.inputDateFrom": "",
        "birthRegistrationForm.inputDateTo": "",
        "birthRegistrationForm.numberNo": "",
        "birthRegistrationForm.bookNo": f"01/{year}",
        "birthRegistrationForm.pageNo": "",
        "birthRegistrationForm.rpFullName": "",
        "birthRegistrationForm.rpBirthDate": "",
        "birthRegistrationForm.rpGender": -2,
        "birthRegistrationForm.registrationDateFrom": "",
        "birthRegistrationForm.registrationDateTo": "",
        "birthRegistrationForm.registrationPlaceStr": -1,
        "q": "*",
        "start": 0,
        "count": 1000,
        "startval": 0
    }

    # Retry mechanism
    retries = 3
    for attempt in range(retries):
        try:
            # Send a POST request with the payload and a longer timeout
            response = requests.post(url, data=payload, cookies=cookies, timeout=30)

            # Check if the request was successful
            if response.status_code == 200:
                try:
                    # Parse the JSON data
                    data = response.json()

                    # Append the data to the list
                    all_data.append(data)
                    break  # Exit the retry loop if successful
                except ValueError:
                    print(f"Response for year {year} is not in JSON format")
                    print(response.text)
                    break  # Exit the retry loop if response is not JSON
            else:
                print(f"Failed to retrieve data for year {year}: {response.status_code}")
                break  # Exit the retry loop if response status is not 200
        except requests.exceptions.Timeout:
            print(f"Timeout occurred for year {year}, attempt {attempt + 1} of {retries}")
            if attempt < retries - 1:
                time.sleep(5)  # Wait for 5 seconds before retrying
            else:
                print(f"Failed to retrieve data for year {year} after {retries} attempts")
        except requests.exceptions.RequestException as e:
            print(f"Request failed for year {year}, attempt {attempt + 1} of {retries}: {e}")
            if attempt < retries - 1:
                time.sleep(5)  # Wait for 5 seconds before retrying
            else:
                print(f"Failed to retrieve data for year {year} after {retries} attempts")

# Write the data to a JSON file
with open("birth_data.json", "w", encoding='utf-8') as f:
    json.dump(all_data, f, ensure_ascii=False, indent=4)

print("Data from 1999 to 2019 has been written to birth_data.json")
```
